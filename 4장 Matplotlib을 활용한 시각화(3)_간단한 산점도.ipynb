```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

# 1. plt.plot 을 사용한 산점도

```python
x = np.linspace(0, 10, 30)
y = np.sin(x)

plt.plot(x, y, 'o', color = 'black')
```

```python
rng = np.random.RandomState(0)
for marker in ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']:
    plt.plot(rng.rand(1), rng.rand(1), marker,
            label = 'marker = "{0}"'.format(marker))
    plt.legend(numpoints = 1)
    plt.xlim(0, 1.5)
```

```python
plt.plot(x, y, '-ok')
```

```python
plt.plot(x, y, '-p', color = 'gray',
        markersize= 15, linewidth = 5,
        markerfacecolor = 'white',
        markeredgecolor = 'gray',
        markeredgewidth = 2)

plt.ylim(-1.2, 1.2)
```

# 2. plt.scatter를 활용한 산점도

```python
plt.scatter(x, y, marker='o');
```

```python
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.3,
           cmap='viridis')
plt.colorbar();
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
features = iris.data.T

plt.scatter(features[0], features[1], alpha=0.2,
           s = 100*features[3], c=iris.target, cmap='viridis')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1]);
```

# 3. plot과 scatter의 차이: 효율성 측면에서 유의할 점
데이터이 양이 적은 경우에는 그렇게 중요하지 않지만, 데이터세트가 수천 개가 넘어가는 경우에는 plt.plot이 plt.scatter보다 확실히 더 효율적이다.
